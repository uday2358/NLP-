````
module_url = "https://tfhub.dev/tensorflow/bert_en_uncased_L-24_H-1024_A-16/1"
bert_layer=hub.KerasLayer(module_url,trainable=True)
````

````
import numpy as np
def bert_encode(texts,tokenizer,max_len=192):
  input_tokens=[]
  input_masks=[]
  segment_ids=[]

  for text in texts:
    text=tokenizer.tokenize(text)
    text=text[:max_len-2]
    input_seq=['[CLS]']+text+['[SEP]']
    pad_len=max_len-len(input_seq)

    tokens=tokenizer.convert_tokens_to_ids(input_seq)
    tokens+=[0]*pad_len
    pad_masks=[1]*len(input_seq)+[0]*pad_len
    seg_ids=[0]*max_len

    input_tokens.append(tokens)
    input_masks.append(pad_masks)
    segment_ids.append(seg_ids)

  return np.array(input_tokens),np.array(input_masks),np.array(segment_ids)
````

````
import tensorflow as tf
from tensorflow.keras.layers import Input,Dense
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam
import tensorflow_hub as hub 


def Build_model(bert_layer,max_len=192):
  all_tokens=Input(shape=(max_len,),dtype=tf.int32,name="input_tokens")
  all_masks=Input(shape=(max_len,),dtype=tf.int32,name="input_masks")
  all_segments=Input(shape=(max_len,),dtype=tf.int32,name="segment_ids")

  pooled_output,sequence_output=bert_layer([all_tokens,all_masks,all_segments])

  model=Model(inputs=[all_tokens,all_masks,all_segments],outputs=[pooled_output,sequence_output])
  model.compile(Adam(lr=1e-5), loss='binary_crossentropy', metrics=['accuracy'])

  return model
````

````

from tokenization import FullTokenizer
vocab_file=bert_layer.resolved_object.vocab_file.asset_path.numpy()
do_lower_case=bert_layer.resolved_object.do_lower_case.numpy()
tokenizer=FullTokenizer(vocab_file,do_lower_case)
````
